```python
from numpy import genfromtxt
from time import time
from datetime import datetime
from sqlalchemy import Column, Integer, Float, Date, String, VARCHAR
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import create_engine
from sqlalchemy.orm import sessionmaker
import csv
import pandas as pd

Base = declarative_base()

class cdc_npao_topics(Base):
    #Tell SQLAlchemy what the table name is and if there's any table-specific arguments it should know about
    __tablename__ = 'cdc_npao_topics'
    __table_args__ = {'sqlite_autoincrement': True}
    #tell SQLAlchemy the name of column and its attributes:
    class_desc = Column(VARCHAR(50))
    topic = Column(VARCHAR(200))
    question = Column(VARCHAR(200))
    short_desc = Column(VARCHAR(50))
    class_id = Column(VARCHAR(10))
    topic_id = Column(VARCHAR(10))
    question_id = Column(Integer, primary_key=True, nullable=False) 

engine = create_engine('sqlite:///cdc_npao_topics.db')
Base.metadata.create_all(engine)
file_name = 'cdc_npao_topics.csv'
df = pd.read_csv(file_name)
df.to_sql(con=engine, index_label='question_id', name=cdc_npao_topics.__tablename__, if_exists='replace')
```

```python

Base = declarative_base()

class cdc_npao(Base):
    #Tell SQLAlchemy what the table name is and if there's any table-specific arguments it should know about
    __tablename__ = 'cdc_npao'
    __table_args__ = {'sqlite_autoincrement': True}
    #tell SQLAlchemy the name of column and its attributes:
    
    npao = Column(Integer, primary_key=True, nullable=False)
    year_start = Column(VARCHAR(50))
    year_end = Column(VARCHAR(50))
    location_abbr = Column(VARCHAR(50))
    location = Column(VARCHAR(50))
    datasource = Column(VARCHAR(50))
    class_desc = Column(VARCHAR(50))
    topic = Column(VARCHAR(50))
    question = Column(VARCHAR(50))
    data_value_unit = Column(VARCHAR(50))
    data_value_type = Column(VARCHAR(50))
    data_value = Column(VARCHAR(50))
    data_value_alt = Column(VARCHAR(50))
    data_value_footnote_value = Column(VARCHAR(50))
    data_value_footnote = Column(VARCHAR(50))
    low_confidence_limit = Column(VARCHAR(50))
    high_comfidence_limit = Column(VARCHAR(50))
    sample_size = Column(VARCHAR(50))
    total = Column(VARCHAR(50))
    age_range = Column(VARCHAR(50))
    education = Column(VARCHAR(50))
    gender = Column(VARCHAR(50))
    income = Column(VARCHAR(50))
    race = Column(VARCHAR(50))
    geo_location = Column(VARCHAR(50))
    class_id = Column(VARCHAR(50))
    topic_id = Column(VARCHAR(50))
    question_id = Column(VARCHAR(50))
    data_value_type_id = Column(VARCHAR(50))
    location_id = Column(VARCHAR(50))
    stratification_category1 = Column(VARCHAR(50))
    stratification1 = Column(VARCHAR(50))
    stratification_category_id1 = Column(VARCHAR(50))
    stratification_id1 = Column(VARCHAR(50))
    


engine = create_engine('sqlite:///cdc_npao.db')
Base.metadata.create_all(engine)
file_name = 'cdc_npao.csv'
df = pd.read_csv(file_name)
df.to_sql(con=engine, index_label='id', name=cdc_npao.__tablename__, if_exists='replace')
```
